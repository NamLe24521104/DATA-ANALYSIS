```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('times_block.csv')
```

```python
data
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['time'] = data['Date'].dt.month
```

```python
data
```

```python
data.drop('Date', inplace = True, axis = 1)
```

```python
data
```

```python
dict_label = {
    "Instagram" : 0,
    "Whatsapp" : 1
}
```

```python
data['App'] = data['App'].map(dict_label)
```

```python
data
```

```python
data_columns = data.columns
```

```python
data_columns
```

```python
fig, axes = plt.subplots(3,2, figsize = (10,8))
axes = axes.flatten()
for i, col in enumerate(data_columns):
    sns.histplot(data[col], ax = axes[i])
plt.tight_layout()
plt.show()
```

```python
def lamda(x):
    if (x == 8):
        return 0
    else:
        return 1
data['time'] = data['time'].apply(lamda)
```

```python
data_normal = ['Usage', 'Notifications', 'Times opened']
```

```python
def normalization(x):
    min_values = data[x].min()
    max_values = data[x].max()
    data[x] = data[x].apply( lambda y: (y - min_values) /(max_values - min_values))
```

```python
for i in data_normal:
    normalization(i)
```

```python
fig, axes = plt.subplots(3,2, figsize = (10,8))
axes = axes.flatten()
for i, col in enumerate(data_columns):
    sns.histplot(data[col], ax = axes[i])
plt.tight_layout()
plt.show()
```

```python
y = data['App']
```

```python
x = data.drop('App', axis = 1)
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y , test_size = 0.3)
```

```python
data
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
confusion_matrix_1 = confusion_matrix(y_test, y_pred)
```

```python
print(confusion_matrix_1)
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(confusion_matrix_1, annot=True, fmt='d', cmap='Reds')
plt.xlabel('D on')
plt.ylabel('Thc t')
plt.title('Confusion Matrix')
plt.show()
```

```python
y_pred_prob = model.predict_proba(X_test)
```

```python
def set_threshold(x):
    if x[1] > 0.6:
        return 1
    else: return 0
```

```python
y_pred_dudoan = [set_threshold(x) for x in y_pred_prob]
```

```python
confusion_matrix_0_6 = confusion_matrix(y_pred_dudoan, y_test)
```

```python
plt.figure(figsize = (5,3))
sns.heatmap(confusion_matrix_0_6, annot = True)
```

```python
def ve_ROC(y_test, y_pred_proba):

    tpr_list = []
    fpr_list = []
    threshold = np.linspace(0,1,50)
    for t in threshold:
        y_dudoan = (y_pred_proba[:, 1] >= t).astype(int)

        tn, fp, fn, tp = confusion_matrix(y_test, y_dudoan).ravel()

        tpr =  tp / (tp + fn)
        fpr = fp / (fp + tp)
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    return tpr_list, fpr_list


```

```python
tpr_list, fpr_list = ve_ROC(y_test, y_pred_prob)
plt.figure(figsize = (5,3))

plt.plot(tpr_list, fpr_list)

plt.show()
```
