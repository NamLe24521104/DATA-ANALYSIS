```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)
```

```python
from sklearn.ensemble import RandomForestClassifier

# 1. Load dữ liệu chuẩn
df_base = pd.read_csv(url)

# 2. Tạo dữ liệu trùng lặp (Duplicates) - Thường không gây hại nếu phân bổ đúng
df_duplicated = pd.concat([df_base, df_base.iloc[[1, 2, 3]]], ignore_index=True)

# 3. Thử nghiệm 1: Để nguyên dữ liệu trùng
X1 = df_duplicated.drop("species", axis=1)
y1 = df_duplicated["species"]
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.3, random_state=42)

model_rf1 = RandomForestClassifier(random_state=42)
model_rf1.fit(X_train1, y_train1)
acc_1 = accuracy_score(y_test1, model_rf1.predict(X_test1))

# 4. Thử nghiệm 2: Xóa trùng lặp (Cleaning cơ bản)
df_cleaned = df_duplicated.drop_duplicates()
X2 = df_cleaned.drop("species", axis=1)
y2 = df_cleaned["species"]
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.3, random_state=42)

model_rf2 = RandomForestClassifier(random_state=42)
model_rf2.fit(X_train2, y_train2)
acc_2 = accuracy_score(y_test2, model_rf2.predict(X_test2))

print(f"Accuracy khi CÓ hàng trùng: {acc_1:.4f}")
print(f"Accuracy sau khi XÓA hàng trùng: {acc_2:.4f}")
```
